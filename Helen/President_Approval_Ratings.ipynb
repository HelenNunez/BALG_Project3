```python
import pandas as pd
import numpy as np
```

```python
DataSet2 = "president_approval_polls.csv"
DataSet2 = pd.read_csv(DataSet2)
DataSet2
```

```python
President_Approvals = DataSet2[["sample_size", "population_full", "methodology", "end_date", "yes", "no"]]
President_Approvals = President_Approvals.rename(columns={"yes": "Approve", "no":"Disapprove"})
President_Approvals
```
